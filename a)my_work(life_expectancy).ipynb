```python
#to convert Status values into 0,1

df2=df.replace(to_replace=["Developing","Developed"],value=[0,1])
df2
```

```python
df2["Status"]=df2["Status"].astype(float)   #values string nn matti number akyale dtype change akan pattu
```

```python
#fill missingvalues through mean

df2["Life expectancy "]=df2["Life expectancy "].fillna(df2["Life expectancy "].mean())
df2["Adult Mortality"]=df2["Adult Mortality"].fillna(df2["Adult Mortality"].mean())
df2["Alcohol"]=df2["Alcohol"].fillna(df2["Alcohol"].mean())
df2["Hepatitis B"]=df2["Hepatitis B"].fillna(df2["Hepatitis B"].mean())
df2[" BMI "]=df2[" BMI "].fillna(df2[" BMI "].mean())
df2["Polio"]=df2["Polio"].fillna(df2["Polio"].mean())
df2["Total expenditure"]=df2["Total expenditure"].fillna(df2["Total expenditure"].mean())
df2["Diphtheria "]=df2["Diphtheria "].fillna(df2["Diphtheria "].mean())
df2["GDP"]=df2["GDP"].fillna(df2["GDP"].mean())
df2["Population"]=df2["Population"].fillna(df2["Population"].mean())
df2[" thinness  1-19 years"]=df2[" thinness  1-19 years"].fillna(df2[" thinness  1-19 years"].mean())
df2[" thinness 5-9 years"]=df2[" thinness 5-9 years"].fillna(df2[" thinness 5-9 years"].mean())
df2["Income composition of resources"]=df2["Income composition of resources"].fillna(df2["Income composition of resources"].mean())
df2["Schooling"]=df2["Schooling"].fillna(df2["Schooling"].mean())
```
